```python
k=2
point1=z[0]
point2=z[1]
list1x=[]
list1y=[]
list2x=[]
list2y=[]
prev1x=0
prev1y=0
prev2x=0
prev2y=0
mean1x=1
mean1y=2
mean2x=1
mean2y=2

while(prev1x!=mean1x and prev1y!=mean1y and prev2x!=mean2x and prev2y!=mean2y):
    list1x=[]
    list1y=[]
    list2x=[]
    list2y=[]
    for i in z:
        prev1x=mean1x
        prev1y=mean1y
        prev2x=mean2x
        prev2y=mean2y
        sub1=i-point1
        dist1=(sub1[0])**2 +(sub1[1])**2
        sub2=i-point2
        dist2=(sub2[0])**2 +(sub2[1])**2
        if(dist1<dist2):
            list1x.append(i[0])
            list1y.append(i[1])
        else:
            list2x.append(i[0])
            list2y.append(i[1])
    mean1x=sum(list1x)/len(list1x)
    mean1y=sum(list1y)/len(list1y)
    mean2x=sum(list2x)/len(list2x)
    mean2y=sum(list2y)/len(list2y)
    point1=[mean1x,mean1y]
    point2=[mean2x,mean2y]
    plt.scatter(list1x,list1y)
    plt.scatter(list2x,list2y)
    plt.scatter(mean1x,mean1y)
    plt.scatter(mean2x,mean2y)
    

```
